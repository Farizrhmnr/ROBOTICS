```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Membuat data dummy untuk pelatihan
np.random.seed(0)
X = np.linspace(0, 10, 100)
y = 2 * X + 1 + np.random.normal(0, 2, size=len(X))
```

```python
# Membuat model neural network sederhana
model = Sequential()
model.add(Dense(units=1, input_dim=1, activation='linear'))
```

```python
# Menentukan optimizer dan loss function
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=optimizer, loss='mean_squared_error')
```

```python
# Melatih model dengan data dummy
model.fit(X, y, epochs=100)
```

```python
# Memprediksi nilai dengan model yang telah dilatih
predictions = model.predict(X)
```

```python
# Menampilkan hasil akhir
plt.scatter(X, y, label='Data Asli')
plt.plot(X, predictions, label='Prediksi', color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
